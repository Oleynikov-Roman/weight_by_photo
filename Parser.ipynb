# Парсер

Все данные были взяты с сайта https://height-weight-chart.com/

Импортируем библиотеки:

```python
import requests 
from bs4 import BeautifulSoup
import urllib
import pandas as pd
import csv
import re
import tqdm
```

Спарсим с главной страницы список ссылок на страницу,на которых находятся фото

```python
base_link='https://height-weight-chart.com/'
respones=requests.get(base_link).text
soup_for_links=BeautifulSoup(respones,'lxml')
links=[]
chart=soup_for_links.find('chart').find_all('a')
for i in chart:
    inn=i.get('href')
    links.append(inn)
print(len(links))
```

```python
links
```

В цикле пройдем по массиву с ссылками,переходя по ссылкам и скачивая фото в папку photo,посредством библиотеки request.Параллельно заносим в таблицу название картинки и вес изображенного на ней человека

```python
number=0
for i in range(len(links)+1):
    try:
        new_link=f'https://height-weight-chart.com/{links[i]}'
        respones_try=requests.get(new_link).text
        soup_try=BeautifulSoup(respones_try,'lxml')
        image=soup_try.find('div',class_='main').find_all('img',class_='largepic')
        weight=re.findall(r".....kg",soup_try.find('p').text)
        for i in image:
            srcc=i.get('src')
            resource = requests.get(f'https://height-weight-chart.com/{srcc}').content
            with open(f'photo/{number}.jpg','wb') as file:
                file.write(resource)
            with open ('photo/data.csv','a') as file:
                writer=csv.writer(file)
                writer.writerow((f'{number}.jpg',weight))
            print(f'Photo :{number}.jpg and weight: {weight}',)
            number+=1
            
    except:
        print(f'\n\t Link {links[i]} not work \n\t')
        i+=1
```

```python
number
```

Формируем финальный файл датасета:

```python
data=pd.read_csv('photo/data.csv',sep=';')
data=data.dropna(axis=0).reset_index(drop=True)
for i in range(data.shape[0]):
    data['weight'][i]=re.findall(r"[0-9]+",data['weight'][i])
    data['weight'][i]=int(data['weight'][i][0])
data['weight']=data['weight'].apply(pd.to_numeric)
print(data.info())
data.to_csv('data.csv',sep=',')
```
